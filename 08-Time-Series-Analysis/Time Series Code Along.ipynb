```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airline = pd.read_csv('airline_passengers.csv', index_col = "Month")
```

```python
airline.dropna(inplace = True)
```

```python
airline.index = pd.to_datetime(airline.index)
```

```python
airline.head()
```

```python
airline['6-SMA'] = airline["Thousands of Passengers"].rolling(6).mean()
airline['12-SMA'] = airline["Thousands of Passengers"].rolling(12).mean()
airline['EWMA-12'] = airline["Thousands of Passengers"].ewm(span = 10).mean()
airline[['EWMA-12', 'Thousands of Passengers']].plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(airline['Thousands of Passengers'], model = 'additive')
```

```python
fig = result.plot()
```
